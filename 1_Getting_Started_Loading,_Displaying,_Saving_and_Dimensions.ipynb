```
import cv2
from matplotlib import pyplot as plt

# Load an image using 'imread' specifying the path to image
image = cv2.imread('/content/maracas-beach.jpg')

# Check if image loading was successful
if image is None:
    print("Error: Image could not be loaded. Check the file path and ensure the image exists.")
else:
    #Show the image with matplotlib
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

Let's create a simple function to make displaying our images simpler and easier

```
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```
# Let's test it out
imshow("Displaying Our First Image", image)
```

Remember Images are arrays:

We need to use numpy to perform this operation. Don't worry, numpy will become one of your best friends if you're learning a Data Science and Computer Vision.

```
# Import numpy
import numpy as np

print(image.shape)
```

```
# To access a dimension, simply index it by using 0, 1 or 2.
image.shape[0]
```

You can see the first dimension is the height and it's 960 pixels

The second dimension is the width which is 1280 pixels.

We can print them out nicely like this:

```
print('Height of Image: {} pixels'.format(int(image.shape[0])))
print('Width of Image: {} pixels'.format(int(image.shape[1])))
print('Depth of Image: {} colors components'.format(int(image.shape[2])))
```

# Print the dimensions of an image loaded using OpenCV (cv2).

It assumes you have an image loaded and stored in a variable called image.

------------------------------------
# Explanation:

#image.shape:
This attribute of the image object returns a tuple containing the dimensions of the image.

The tuple is typically in the order of (height, width, channels).

image.shape[0] represents the height of the image in pixels.

image.shape[1] represents the width of the image in pixels.

image.shape[2] represents the number of color channels in the image.

#int(...):
This function converts the values obtained from image.shape into integers.
This is done to ensure the values are displayed correctly.

#.format(...):
This is a method used for string formatting. It replaces the {} placeholder in the string with the value provided within the format() method.

------------------------
#Line-by-line breakdown:

#print('Height of Image: {} pixels'.format(int(image.shape[0]))):

This line prints the height of the image.

It gets the height value from image.shape[0], converts it to an integer, and then inserts it into the string "Height of

Image: {} pixels" using the format() method. The resulting string is then printed to the console.

print('Width of Image: {} pixels'.format(int(image.shape[1]))):

This line does the same as above, but for the width of the image.

print('Depth of Image: {} colors components'.format(int(image.shape[2]))): This line prints the number of color channels (depth) of the image.

For example, a color image usually has 3 channels (Red, Green, Blue - RGB), while a grayscale image has only 1 channel.

----------------------------
In summary:

The code snippet prints the dimensions of the loaded image in the following format:


Height of Image: [height value] pixels

Width of Image: [width value] pixels

Depth of Image: [number of color channels] colors components


For example, if the image has a height of 480 pixels, a width of 640 pixels, and 3 color channels, the output would be:

Height of Image: 480 pixels

Width of Image: 640 pixels

Depth of Image: 3 colors components
